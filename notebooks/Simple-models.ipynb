```python
def extract(strings):
    """one-time use extractor"""
    mae=0
    mse=0
    mape=0
    r2=0
    spear=0
    kendall=0
    cindex=0
    vals=['MAE','MAPE','MSE','R2','Spear','Kendall','C-index']

    numbers=np.zeros((len(strings),7))
    for i,string in enumerate(strings):
        for ii,line in enumerate(string.split('\n')):
            line=line.split(' ')
            numbers[i,ii]=float(line[-1])
    means = np.mean(numbers,axis=0)
    stds = np.std(numbers,axis=0)

    output=[]
    for name, m,s in zip(vals,means, stds):
       string= f'{name}: {m}, [{s}]'
       output.append(string)
    return output     
```
